```python
def which_prize(points):
    if points in range(0,51):
        prize_name = "wooden rabit"
    elif points in range(51,151):
        return "oh dear, no prize this time."
    elif points in range(151, 181):
        prize_name = "water-thin mint"
    elif points in range(181,201):
        prize_name = "penguin"
    
    message = 'Congratulations! you have won a {}'.format(prize_name)
    return (message)

which_prize(181)
```
